```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/bank.csv")

print(df)
df.info()
print(df.isnull().sum())

# -----------------------------
# FIX: Only fill missing values
# for columns that actually exist
# -----------------------------

# 1. age â†’ numeric â†’ median
if 'age' in df.columns:
    df['age'] = df['age'].fillna(df['age'].median())

# 2. job â†’ categorical â†’ mode
if 'job' in df.columns:
    df['job'] = df['job'].fillna(df['job'].mode()[0])

# 3. marital â†’ categorical â†’ mode
if 'marital' in df.columns:
    df['marital'] = df['marital'].fillna(df['marital'].mode()[0])

# 4. education â†’ categorical â†’ mode
if 'education' in df.columns:
    df['education'] = df['education'].fillna(df['education'].mode()[0])

# 5. default â†’ categorical â†’ mode
if 'default' in df.columns:
    df['default'] = df['default'].fillna(df['default'].mode()[0])

# 6. balance â†’ numeric â†’ median (skewed)
if 'balance' in df.columns:
    df['balance'] = df['balance'].fillna(df['balance'].median())

# 7. housing â†’ categorical â†’ mode
if 'housing' in df.columns:
    df['housing'] = df['housing'].fillna(df['housing'].mode()[0])

# 8. loan â†’ categorical â†’ mode
if 'loan' in df.columns:
    df['loan'] = df['loan'].fillna(df['loan'].mode()[0])

# 9. contact â†’ categorical â†’ mode
if 'contact' in df.columns:
    df['contact'] = df['contact'].fillna(df['contact'].mode()[0])

# 10. duration â†’ numeric â†’ mean
if 'duration' in df.columns:
    df['duration'] = df['duration'].fillna(df['duration'].mean())

# 11. campaign â†’ numeric â†’ median
if 'campaign' in df.columns:
    df['campaign'] = df['campaign'].fillna(df['campaign'].median())

# 12. poutcome â†’ categorical â†’ mode
if 'poutcome' in df.columns:
    df['poutcome'] = df['poutcome'].fillna(df['poutcome'].mode()[0])

# Show final missing values
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
# Check total number of duplicate rows
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)

#remove duplicates
df = df.drop_duplicates()
```

```python
#Problem 5
# ----------------------------
# Titanic Data Cleaning & Visualization
# ----------------------------

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv")

# Step 2: Optional â€“ inspect the dataset
print(df.head())
print(df.info())
print(df['Embarked'].unique())

# Step 3: Drop rows with missing 'Embarked' (if any)
df = df.dropna(subset=['Embarked'])

# Step 4: One-hot encoding of 'Embarked' (optional)
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked')
df = pd.concat([df, embarked_dummies], axis=1)

# Step 5: Check missing values in 'Age' and drop if needed
print("Missing Age values:", df['Age'].isnull().sum())
df = df.dropna(subset=['Age'])  # only keep rows with Age

# Step 6: Visualize survival rate by gender
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Mean Survival Rate by Gender")
plt.ylabel("Mean Survival Rate")
plt.show()

# Step 7: Visualize survival rate by gender AND port of embarkation
plt.figure(figsize=(8,5))
sns.barplot(x='Embarked', y='Survived', hue='Sex', data=df)
plt.title("Survival Rate by Gender and Port of Embarkation")
plt.ylabel("Mean Survival Rate")
plt.show()

# Step 8: Optional â€“ count plot for absolute numbers
plt.figure(figsize=(8,5))
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title("Survival Counts by Port of Embarkation")
plt.show()
```
